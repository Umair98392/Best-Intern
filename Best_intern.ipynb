# Perform Text Preprocessing

```python
import re
def text_preprocessing(text):
    text = re.sub(r"\(.*\)", '', text)
    text = re.sub(r"-", '', text)
    text = re.sub(r"\.", '', text)
    return text
data['Other skills'] = data['Other skills'].apply(text_preprocessing)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
```

```python
data.head()
```

# Drop Column 'index'

```python
data.drop(['index'],inplace=True,axis=1)
```

# Seperate skills in Other skills column

```python
# Import the necessary libraries
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Extract unique skills from the "other skills" column
unique_skills = set()
for skills in data['Other skills']:
    skills_list = skills.split(',')  # Split the skills by comma or any other delimiter
    unique_skills.update(skills_list)

# Create a binary feature matrix for the unique skills
mlb = MultiLabelBinarizer()
skills_matrix = mlb.fit_transform(data['Other skills'].str.split(','))

# Create a dataframe with the skills matrix
skills_df = pd.DataFrame(skills_matrix, columns=mlb.classes_)


# Concatenate the skills dataframe with the original dataset
data = pd.concat([data, skills_df], axis=1)

# Drop the original "other skills" column
data.drop('Other skills', axis=1, inplace=True)

# Print the updated dataset

```
